### Функции

````python
# преобразует юникод в символ
print(chr(100))
# преобразует символ в юникод
print(ord('d'))

s = 'python'

print(s[2:4])
print(s[::2])

# f строки
x = 4
y = 6
z = x * y
print(f'{x} + {y} = {z}')

# замена символов
s = 'python'
s = s.replace('h', 't')
print(s)

s = s.capitalize()
print(s)
s = s.upper()
print(s)
s = s.lower()
print(s)
s = s.swapcase()
print(s)
title = 'the sun is rising'
title = title.title() # все первые буквы слов - заглавные
print(title)
````

### Подсчет и совпадения

````python
s = 'foo moo goo'
print(s.count('oo'))
print(s.count('oo', 0, 8))
print(s.endswith('goo', 0, 8))
print(s.endswith('goo'))
print(s.startswith('moo'))
print(s.startswith('moo', 4))

print(s.find('oo')) # возвращает индекс начала заданной подстроки. если ее нет, то -1
print(s.rfind('oo')) # тоже самое, но возврашает последний индекс подстроки или -1
print(s.index('oo')) # то же, что и find, но возвращает Value Error
# print(s.index('lua'))
print(s.rindex('oo'))
````

### Классификация

````python
s = 'abc123'
not_s = 'abc$123'
# состоит ли строка только из букв и цифр
print(s.isalnum(), not_s.isalnum())
print(s[0:3].isalpha())
print(s[4:5].isdigit())

print(s.islower())
print(s.isupper())
````

### Форматирование строк

````python
s = 'python'
s = s.center(10, '/') # выравнивает строку до заданной ширины по центру(пробелами по умолчанию)
print(s)

s = 'py\tth\ton'
s = s.expandtabs(3) # замена табуляции пробелами(8 штук по умолчанию)
print(s)

s = s.ljust(15, '-') # выравнивание по левому краю
print(s)
s = s.rjust(20, '-')
print(s)

s_copy = s

s = s.lstrip('-') # убирает все символы слева
print(s)
s = s.rstrip('-')

s = s_copy.strip('-') # убирает с обеих сторон
print(s)

s = 'I hate Python I love Java'
s = s.replace('hate', 'love') # можно указать колво раз
print(s)

print('51'.zfill(8)) # дополняет нулями до нужной длины
print('+51'.zfill(8))
````

### Строки в списки и наоборот

````python
sample_list = ['foo', 'bar', 'far', 'boo']
s = '-'.join(sample_list)
print(s)

print(':'.join('hello'))

print('hey|python'.partition('|'))
# но
print('hey|python|foo'.partition('|'))

print('foo, bar, boo'.split(', ', 1))
print('foo, bar, boo'.rsplit(', ', 1))
````

| Разделитель | Значение |
|------------|----------|
| `\n` | Новая строка |
| `\r` | Возврат каретки |
| `\r\n` | Возврат каретки + перевод строки |
| `\v` или `\x0b` | Вертикальная табуляция (таблицы строк) |
| `\f` или `\x0c` | Подача формы |
| `\x1c` | Разделитель файлов |
| `\x1d` | Разделитель групп |
| `\x1e` | Разделитель записей |
| `\x85` | Следующая строка |
| `\u2028` | Новая строка (Unicode) |
| `\u2029` | Новый абзац (Unicode) |


````python
print('foo\nbar\r\nbaz\fqux\u2028quux'.splitlines()) # делит строку на подстроки
````

8. Дана строка:

text = "  Data Science, AI & ML!!!  "


Сделай строку вида:

"data science ai ml"


(подсказка: lower, replace, strip, split, join)

````python
text = "  Data Science, AI & ML!!!  "

new_text = ' '.join(text.lower().replace(',', '').replace('&', '').replace('!', '').split())
print(new_text)
````

9. Дан список строк:

texts = [
    "  Hello World! ",
    "Python   is GREAT",
    "  AI & Data Science  "
]


Создай новый список, где каждая строка:

в нижнем регистре

без лишних пробелов

без ! и &

````python
texts = [
    "  Hello World! ",
    "Python   is GREAT",
    "  AI & Data Science  "
]

new_texts = [' '.join(n.lower().replace('!', '').replace('&', '').strip().split()) for n in texts]
print(new_texts)
````
