4. –°–æ–∑–¥–∞–π —Ñ–∞–π–ª people.csv —Å —Å–æ–¥–µ—Ä–∂–∏–º—ã–º:

name,age,city
Alice,25,Berlin
Bob,17,Paris
Charlie,30,Berlin

–ù–∞–ø–∏—à–∏ –∫–æ–¥, –∫–æ—Ç–æ—Ä—ã–π:

—á–∏—Ç–∞–µ—Ç —Ñ–∞–π–ª

–ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç –∑–∞–≥–æ–ª–æ–≤–æ–∫

—Å–æ–∑–¥–∞—ë—Ç —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤–∞—Ä–µ–π –≤–∏–¥–∞:

[
  {"name": "Alice", "age": 25, "city": "Berlin"},
  ...
]

````python
with open('files/people.csv') as csv_file:
    header = csv_file.readline().strip().split(',')
    lines = csv_file.readlines()

data = []
for i in range(len(lines)):
    line = lines[i].strip().split(',')
    data.append(dict([(header[h], line[h]) for h in range(len(header))]))

print(data)
````

5. –ò—Å–ø–æ–ª—å–∑—É—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∏–∑ –∑–∞–¥–∞—á–∏ 4:

–≤—ã–≤–µ–¥–∏ –∏—Ö –∏–º–µ–Ω–∞

````python
print([l['name'] for l in data if int(l['age']) > 18])
````

6. –î–∞–Ω —Ñ–∞–π–ª numbers.txt (–∏–∑ –∑–∞–¥–∞—á–∏ 3).

–ù–∞–ø–∏—à–∏ –∫–æ–¥, –∫–æ—Ç–æ—Ä—ã–π:

—á–∏—Ç–∞–µ—Ç —á–∏—Å–ª–∞ –∏–∑ —Ñ–∞–π–ª–∞

–Ω–∞—Ö–æ–¥–∏—Ç —Å—É–º–º—É, –º–∏–Ω–∏–º—É–º –∏ –º–∞–∫—Å–∏–º—É–º

–≤—ã–≤–æ–¥–∏—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ —Ñ–æ—Ä–º–∞—Ç–µ:

sum=15, min=1, max=5

````python
with open('files/numbers.txt') as f:
    lines = []
    read_result = f.readline()
    while read_result != '':
        lines.append(int(read_result.strip()))
        read_result = f.readline()

print(f'sum={sum(lines)}, min={min(lines)}, max={max(lines)}')
````

1. –ß—Ç–µ–Ω–∏–µ CSV –≤ —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤–∞—Ä–µ–π

–î–∞–Ω —Ñ–∞–π–ª employees.csv:

id,name,department,salary
1,Alice,IT,5000
2,Bob,HR,3500
3,Charlie,IT,6000
4,Diana,Marketing,4000

–ó–∞–¥–∞—á–∞:

–ø—Ä–æ—á–∏—Ç–∞—Ç—å —Ñ–∞–π–ª

—Å–æ–∑–¥–∞—Ç—å —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤–∞—Ä–µ–π

id –∏ salary –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ int

üëâ –†–µ–∑—É–ª—å—Ç–∞—Ç:

[
  {"id": 1, "name": "Alice", "department": "IT", "salary": 5000},
  ...
]

````python
import csv

with open('files/employees.csv', newline='') as f:
    d_reader = csv.DictReader(f)

    employees_dict = []
    header = d_reader.fieldnames
    reader = d_reader.reader
    for row in reader:
       for i in range(len(row)):
           try:
               row[i] = int(row[i])
           except TypeError, ValueError:
               print()

       employees_dict.append()



print(employees_dict)

````
