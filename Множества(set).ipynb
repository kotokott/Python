# Множества

### Создание

```python
s1 = set()
s2 = set({1, 'python', True})
s3 = {2, 'Java', False}

# нельзя добавлять в множество(сет) изменяемые объекты
# error_s = {[1, 2, 3, 4], 20.4}

print(s3)
```

### Взаимодействие

```python
s3 = {2, 'Java', False}

# добавление
s3.add(20.6)

s3.update([4, 8, 'C++'])

# удаление
s3.remove(4)

# вернет ошибку, так как элемента нет в сете
# s3.remove(14)

# то же, что и remove, но без ошибок
s3.discard(14)

# удаляет случайный элемент
s3.pop()
```

### Слияние, пересечение и обновление

```python
a = {1, 2, 3}
b = {2, 3, 4, 5}

# объединение
c = a | b
print(c)

# пересечение
c = a & b
print(c)

# разность
c = a - b
print(c)

# симметричная разность
c = a ^ b
print(c)

# подмножество и надмножество
a = {3, 4}
print(a <= b)
print(b >= a)
```

5. Даны два списка:

a = [1, 2, 3, 4, 5]
b = [4, 5, 6, 7]


С помощью множеств найди:

общие элементы

элементы, которые есть в a, но нет в b

все уникальные элементы из обоих списков

```python
a = [1, 2, 3, 4, 5]
b = [4, 5, 6, 7]
a = set(a)
b = set(b)

c = a & b
print(c)

c = a - b
print(c)

c = a | b
print(c)
```

tags = ["ml", "python", "ai", "ml", "data", "python", "ai"]
