## Массивы(Ndarray)

Основной элемент библиотеки NumPy — объект ndarray (что значит N-размерный массив). Этот объект является многомерным однородным массивом с заранее заданным количеством элементов. Однородный — потому что практически все объекты в нем одного размера или типа. На самом деле, тип данных определен другим объектом NumPy, который называется dtype (тип-данных). Каждый ndarray ассоциирован только с одним типом dtype.

Количество размерностей и объектов массива определяются его размерностью (shape), кортежем N-положительных целых чисел. Они указывают размер каждой размерности. Размерности определяются как оси, а количество осей — как ранг.

### Создание

```python
import numpy as np

# одномерный массив
a = np.array([1, 2, 3])
print(a, a.dtype)

'''
Только что созданный массив имеет одну ось,
а его ранг равняется 1, то есть его форма — (3,1)
'''

print(a.ndim) # оси
print(a.size) # размер
print(a.shape) # форма - кортеж из прошлых двух показателей
```

```python
import numpy as np

# двумерный массив
b = np.array([[1.3, 2.5], [0.4, 3.1]])
print(b, b.dtype)

print(b.shape) # два измерения по два элемента в каждом
```

```python
d = np.array(((1, 2, 3), (5, 4, 2)))
print(d)

print()
e = np.array([(1, 2, 3), [5, 6, 3], (7, 4, 1)])
print(e)
```

### Типы данных

Массивы NumPy сделаны так, чтобы включать самые разные типы данных. Например, можно включать строки:

| **int** | Целое число по умолчанию | Аналог long в C; обычно int64 или int32 |
| **intc** | Целое число (C-style) | Идентичный int в C; обычно int32 или int64 |
| **intp** | Индексное целое число | Аналог size_t в C; обычно int32 или int64 |
| **int8** | Байт | от -128 до 127 |
| **int16** | Целое число | от -32768 до 32767 |
| **int32** | Целое число | от -2147483648 до 2147483647 |
| **int64** | Целое число | от -9223372036854775808 до 9223372036854775807 |
| **uint8** | Беззнаковое целое | от 0 до 255 |
| **uint16** | Беззнаковое целое | от 0 до 65535 |
| **uint32** | Беззнаковое целое | от 0 до 4294967295 |
| **uint64** | Беззнаковое целое | от 0 до 18446744073709551615 |
| **float** | Число с плавающей точкой | Псевдоним для float64 |

| **float32** | Одинарная точность | 8-бит экспонента, 23-бит мантисса |
| **float64** | Двойная точность | 11-бит экспонента, 52-бит мантисса |
| **complex** | Комплексное число | Псевдоним для complex128 |
| **complex64** | Комплексное число | Два float32 (реальная и мнимая части) |
| **complex128** | Комплексное число | Два float64 (реальная и мнимая части) |


```python
f = np.array([[1, 2, 3], [4, 5, 6]], dtype=complex)
print(f.dtype)
```

### Функции генерации массива

Библиотека NumPy предоставляет набор функций, которые генерируют ndarray с начальным содержимым. Они создаются с разным значениями в зависимости от функции. Это очень полезная особенность. С помощью всего одной строки кода можно сгенерировать большой объем данных.

```python
import numpy as np

# создание массива из нулей с заданными размерностями
g = np.zeros((4, 4), dtype=np.int8)
print(g)
h = np.ones((3, 3), dtype=np.int8) # тоже самое, но с еденицами
print(h)

i = np.arange(2, 10) # генерация массива с числами в заданном диапазоне
print('\n', i)
j = np.arange(1, 15, 3) # третий параметр отвечает за шаг
print(j)

k = np.arange(4, 20, 2).reshape(4, 2) # reshape отвечает за размерность массива
print('\n', k)

l = np.linspace(0, 10, 17) #разбивает последовательность на n чисел(третий аргумент)
print('\n', l)

m = np.random.random(3) # генерация случайных чисел
print('\n', m)
n = np.random.random((3, 3))
print(n)
```

## Операции

### Арифметические

```python
a = np.arange(4)
print(a)
print(a + 4)
print(a * 2)
# и тд

b = np.arange(4, 8)
print('\n', a, b)
print(a + b)
print(a - b)
print(a * b)

print('\n', a * np.sin(b))
```

```python
c = np.arange(4, 10).reshape(2, 3)
d = np.ones((2, 3))
print(c)
print(d)
print(c + d * 2)
```

### Произведение матриц

Выбор оператора для поэлементного применения — это странный аспект работы с библиотекой NumPy. В большинстве инструментов для анализа данных оператор * обозначает произведение матриц. Он применяется к обоим массивам. В NumPy же подобное произведение обозначается функцией dot(). Эта операция не поэлементная.

image.png

```python
a = np.arange(1, 16, 2).reshape(2, 4)
b = np.arange(2, 10).reshape(4, 2)
c = np.dot(a, b)
print(c)
print(a.dot(b))
print(b.dot(a)) # не коммутативная операция
```

### Инкремент и декремент

```python
a = np.arange(4)
print(a)
a += 1
print(a)
a -= 1
print(a)
```

### Универсальные функции(ufunc)

Универсальная функция, известная также как ufunc, — это функция, которая применяется в массиве к каждому элементу.

```python
a = np.arange(1, 5)
print(np.log(a))
print(np.sqrt(a))
print(np.sin(a))
```

### Функции агрегации

Функции агрегации выполняют операцию на наборе значений, например, на массиве, и выдают один результат. Таким образом, сумма всех элементов массива — это результат работы функции агрегации.

```python
a = np.arange(1, 5)
print(a.sum())
print(a.min())
print(a.max())
print(a.mean()) # среднее арифметическое
print(a.std())
'''
Вычисляет стандартное отклонение (Standard Deviation).
Простыми словами: это показатель того,
насколько сильно числа в массиве «разбросаны» относительно их среднего арифметического
'''
```
