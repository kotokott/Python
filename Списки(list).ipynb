### Вложенные списки

```python
listWithLists = [[1,2], [3, 4], [5, 6], [7, 8], [9, 0]]

matrix = [
[1, 2, 3],
[4, 5, 6],
[7, 8, 9]
]

print(matrix[1][0])

for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        print(matrix[i][j], end=' ')


# рекурсивная обработка матриц
def recursive_matrix_processing(matrix, func= print):
    for i in matrix:
        if isinstance(i, list):
            recursive_matrix_processing(i, func)
        else:
            func(i)

print('\n')
recursive_matrix_processing(matrix)
```

### List comprehension - **Важно**

```python
# буквально генератор списков
# быстрее for

print('nums')
nums = [n for n in range(1, 6)]
print(nums)

print('squares')
squares = [n*n for n in nums]
print(squares)

print('even')
even_squares = [n*n for n in nums if n % 2 == 0]
print(even_squares)

print('matrix')
matrix = [
    [x for x in range(1, 4)]
    for y in range(1, 4)
]
for i in range(len(matrix)):
    print(matrix[i])
```

5. Дан список чисел:

nums = list(range(1, 21))


С помощью list comprehension:

получи список квадратов чётных чисел

получи список чисел, которые делятся на 3, но не на 5

```python
nums = list(range(1, 21))

squares = [n*n for n in nums if n % 2 == 0]
print(squares)

# n % 5 == 1 - это значит остаток равен 1, а не не делится на 5
specificNums = [n for n in nums if n % 3 == 0 and n % 5 != 0]
print(specificNums)
```

6. Дан список слов:

words = ["data", "science", "ai", "model", "python"]


Создай список:

слов длиной больше 3

слов, приведённых к верхнему регистру, но только если длина ≥ 4

```python
words = ["data", "science", "ai", "model", "python"]

longWords = [w for w in words if len(w) > 3]
print(longWords)

HIGHWords = [w.upper() for w in words if len(w) >= 4]
print(HIGHWords)
```
