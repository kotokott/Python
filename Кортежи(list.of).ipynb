# Кортежи

_Основное отличие от списков в том, что их содержимое нельзя изменить_
по этой причине, они быстрее списков

### Создание

```python
cake = ('c', 'a', 'k', 'e')
print(cake)

cookie = ('c', 0, 0, 'k', 'i', 'e')
print(''.join([str(n) for n in cookie]))

nums = 1, 2, 3, 4, 5
```

### Взаимодействие

```python
# элементы неизменяемы
cookie = ('c', 0, 0, 'k', 'i', 'e')
# cookie[1] = 'o'

# cookie.remove('c')

# итерируемые ли объекты?
a = (1, 2)
print(any(a))
```

### Особенности

```python
nums = (1, 2, 2, [3, 4])

# индексы одинаковых элементов равны
print(id(nums[1]) == id(nums[2]))
print(id(nums[1]) == id(nums[3]))

# но
nums[3].append(14)
print(nums)

# присваивание элементам одного кортежа, значений другого
a = 1, 2, 3
(one, two, three) = a
print(one, two, three)

person = ("Alice", 25, "Berlin")
name, age, city = person
age = 26

print(name, age, city)
```

9. Дан список чисел:

nums = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]


Создай словарь вида:

{1: 1, 2: 2, 3: 3, 4: 4}


(ключ — число, значение — сколько раз встречается)

```python
nums = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

nums_dict = {}
for n in nums:
    nums_dict[n] = nums_dict.get(n, 0) + 1

print(nums_dict)
```
