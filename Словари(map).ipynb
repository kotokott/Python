### Взаимодействие

```python
dict_sample = {
  "Company": "Toyota",
  "model": "Premio",
  "year": 2012
}
# поиск
x = dict_sample["model"]
print(x)

print(dict_sample.get('year'))

engine = dict_sample.get('engine', 'there is no engine')
print(engine)

# добавление
dict_sample['Capacity'] = '18000C'

dict_sample['Colors'] = 'blue', 'white', 'yellow' # множество set

print(dict_sample)

# удаление
del dict_sample['year']

dict_sample.pop('Capacity')

dict_sample.popitem() # удаление последнего элемента

print(dict_sample)

dict_sample.clear() # очистка
print(dict_sample)

del dict_sample # удаление целого словаря
# print(dict_sample)
```

### Методы

```python
dict_sample = {
  "Company": "Toyota",
  "model": "Premio",
  "year": 2012
}

# если ключа нет в словаре, метод добавит его, если же ключ есть, то ничего не произойдет
x = dict_sample.setdefault('color', 'gray')
print(x)
print(dict_sample)

y = dict_sample.setdefault('model', 'Allion')
print(y)
print(dict_sample)

# возвращает итерируемый объект ключей
dict_keys = dict_sample.keys()
print(dict_keys)

# возвращает ключи с min и max значениями
num_dict = {
    'Max': 14,
    'Tim': 48,
    'Bob': 34
}

min_val = min(num_dict)
max_val = max(num_dict)
print(min_val, max_val)
```

### Слияние и обновление словарей

```python
dict1 = {'x':1, 'y':2}
dict2 = {'a':11, 'b':22}

dict3 = dict1 | dict2
print(dict3)

dict1 |= dict2
print(dict2)

# если в словарях будут пересекающиеся ключи, то останется ключ второго словаря, а первого заменится

dict2['x'] = 5
dict1 |= dict2
print(dict1)
```

### List comprehensions для словарей

```python
people = [{
  "first_name": "Василий",
  "last_name": "Марков",
  "birthday": "9/25/1984"
}, {
  "first_name": "Регина",
  "last_name": "Павленко",
  "birthday": "8/21/1995"
}]

birthdays = [
  person[term]
  for person in people
  for term in person
  if term == "birthday"
]

print(birthdays)
```

1. Дан список слов:

words = ["apple", "banana", "apple", "orange", "banana", "apple"]


Создай словарь, где:

ключ — слово

значение — сколько раз оно встречается

```python
words = ["apple", "banana", "apple", "orange", "banana", "apple"]

words_dict = {}
for w in words:
    words_dict[w] = words_dict.get(w, 0) + 1
    # O(n^2) - сложность строки ниже - неэффективно
    #words_dict[w] = words.count(w)

print(words_dict)
```

3. Дан словарь с оценками:

scores = {
    "math": 90,
    "physics": 85,
    "english": 78,
    "biology": 88
}


выведи только названия предметов

выведи только оценки

выведи пары предмет — оценка

найди предмет с максимальной оценкой

```python
scores = {
    "math": 90,
    "physics": 85,
    "english": 78,
    "biology": 88
}

keys = list(scores.keys())
values = list(map(str, scores.values()))
pairs = [str(k + ':' + str(v)) for k, v in scores.items()]
max_subject = max(scores, key=scores.get)
# поиск не по ключу, а по значению
# max_subject = max(scores)
print(' '.join(keys))
print(' '.join(values))
print(' ' .join(pairs))
print(max_subject)
```

4. Дан словарь пользователей:

users = {
    "u1": {"age": 25, "city": "Berlin"},
    "u2": {"age": 17, "city": "Paris"},
    "u3": {"age": 30, "city": "Berlin"},
    "u4": {"age": 22, "city": "Rome"}
}


Создай список:

id пользователей старше 18

id пользователей из города "Berlin"

```python
users = {
    "u1": {"age": 25, "city": "Berlin"},
    "u2": {"age": 17, "city": "Paris"},
    "u3": {"age": 30, "city": "Berlin"},
    "u4": {"age": 22, "city": "Rome"}
}

major_people_ids = [k for k, v in users.items() if v['age'] > 18]
print(major_people_ids)
berlin_residents = [k for k, v in users.items() if v['city'] == 'Berlin']
print(berlin_residents)
```
